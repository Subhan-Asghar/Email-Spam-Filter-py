```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression 
```

```python
dataset=pd.read_csv(r'D:\Projects\Python\8.Spam-Email-Filter/mail_data.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.loc[dataset['Category']=='spam','Category',]=0
```

```python
dataset.loc[dataset['Category']=='ham','Category',]=1
```

```python
dataset.head()
```

```python
X=dataset['Message']
Y=dataset['Category']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
model=LogisticRegression()
```

```python
feature=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
```

```python
X_train_feature=feature.fit_transform(X_train)
X_test_feature=feature.transform(X_test)

Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
model.fit(X_train_feature,Y_train)
```

```python
print(X_test_feature)
```

```python
prediction=model.predict(X_test_feature)
```

```python
print(prediction)
```

```python
accuracy=accuracy_score(Y_test,prediction)
print(accuracy)
```
